```python
# To han dpick a particular columns into a new dataframe
house_ds_2 = house_ds[[
    'LotArea',
    'BedroomAbvGr',
    'SalePrice'
]]
house_ds_2.head()
```

```python
house_ds_2.max()
```

```python
house_ds_2.min()
```

```python
house_ds_norm1 = house_ds_2 / house_ds_2.max()
house_ds_norm1
```

```python
house_ds_norm1.max()
```

```python
house_ds_mean_norm1 = (house_ds_2 - house_ds_2.mean() / house_ds_2.max() - house_ds_2.min())
house_ds_mean_norm1
```

```python
house_ds_mean_norm2 = (house_ds_2 - house_ds_2.mean()) / house_ds_2.std()
```

```python
house_ds_mean_norm2
```

```python
house_ds_mean_norm2.to_excel('house_mean_normalized.xlsx', index=False, header=True)
```
